```python
from __future__ import print_function
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
pd.options.display.max_columns = 200
pd.options.display.max_rows = 1000
```

# data

```python
train = pd.read_csv("../Data/blogData_train.csv",header=None)
test = pd.read_csv("../Data/blogData_test.csv",header=None)

# rename
names = ["V%i" % i if i != 281 else "y" for i in range(1,282)]
train.columns = names
test.columns = names
```

# preprocessing

```python
X_train = train.ix[:,"V1":"V280"]
y_train = train["y"]

X_test = test.ix[:,"V1":"V280"]
y_test = test["y"]

y_train_log = np.log(train["y"]+1)
```

```python
# create class variables 
y_train_class = np.array([1 if x > 5 else 0 for x in y_train])
y_test_class = np.array([1 if x > 5 else 0 for x in y_test])
```

# evaluate function

```python
def eva(test_pred, test_real):
    return ((np.log(1+test_real)-test_pred)**2).mean()
```

# Two step model

Steps:
    1. gradinet boosting trees for classification
    2. other models for regression on the positive part

## Gradient boosting tree classification

```python
from sklearn.metrics import confusion_matrix
import xgboost as xgb

RANDOM_STATE = 42

params = {
    'n_estimators':1500,
    'min_child_weight': 1,
    'learning_rate': 0.01,
    'colsample_bytree': 0.5,
    'max_depth': 12,
    'subsample': 0.7,
    'reg_alpha': 1,
    'gamma': 0.1,
    'objective':'binary:logistic',
    'silent':True,
    'seed': RANDOM_STATE
}

# xgtrain = xgb.DMatrix(X, label=loss_boxcox[0])
xgtrain = xgb.DMatrix(X_train, label=y_train_class)
xgtest = xgb.DMatrix(X_test)
gb_model = xgb.XGBClassifier(**params)
```

```python
gb_model.fit(X_train, y_train_class)
```

```python
for nt in [100,300,500,600,700,800,900,1000,1100,1200,1300,1400,1500]:
    gb_pred = gb_model.predict(X_test, ntree_limit=nt)
    fpr, tpr, thresholds = metrics.roc_curve(y_test_class, gb_pred)
    print(metrics.auc(fpr, tpr))
```

## Gradient boosting tree regression

```python
def mse(preds, dtrain):
    labels = dtrain.get_label()
    grad = preds - labels
    hess = np.array([1]*len(grad))
    return grad, hess

RANDOM_STATE = 42
params = {
    'min_child_weight': 1,
    'eta': 0.01,
    'colsample_bytree': 0.5,
    'max_depth': 12,
    'subsample': 0.5,
    'reg_alpha': 1,
    'gamma': 0.01,
    'silent':True,
    'seed': RANDOM_STATE
}

# xgtrain = xgb.DMatrix(X, label=loss_boxcox[0])
xgtrain_positive = xgb.DMatrix(X_train.ix[y_train_class==1,:], label=y_train_log[y_train_class==1])
xgtest_positive = xgb.DMatrix(X_test.ix[gb_pred==1,:])
```

```python
gb_model_positive = xgb.train(params,xgtrain_positive,1500,obj = mse)
```

```python
tree_num = [100,200,300,400,500,600,700,800,900,1000,1100,1200,1300,1400,1500]
```

```python
gb_pred_positive = []
```

```python
for nt in tree_num:
    gb_pred_positive.append(gb_model_positive.predict(xgtest_positive,ntree_limit=nt))
```

```python
for pred in gb_pred_positive:
    print(eva(pred, y_test[gb_pred==1]))
```
