# Data

```python
train = pd.read_csv("../Data/blogData_train.csv",header=None)
test = pd.read_csv("../Data/blogData_test.csv",header=None)

# rename
names = ["V%i" % i if i != 281 else "y" for i in range(1,282)]
train.columns = names
test.columns = names

X_train = train.ix[:,"V1":"V280"]
y_train = train["y"]

X_test = test.ix[:,"V1":"V280"]
y_test = test["y"]

y_train_log = np.log(train["y"]+1)
y_test_log = np.log(test["y"]+1)

y_train_class = np.array([1 if x > 0 else 0 for x in y_train]).reshape(-1)
y_test_class = np.array([1 if x > 0 else 0 for x in y_test]).reshape(-1)

def eva(test_pred, test_real):
    return ((test_real-test_pred)**2).mean()
```

```python
y_train_class =  np.array([1 if x > 0 else 0 for x in y_train]).reshape(-1)
y_test_class = np.array([1 if x > 0 else 0 for x in y_test]).reshape(-1)
```

# Gradient Boosting Model

```python
import xgboost as xgb

def mse(preds, dtrain):
    labels = dtrain.get_label()
    grad = preds - labels
    hess = np.array([1]*len(grad))
    return grad, hess
    
RANDOM_STATE = 42
params = {
    'min_child_weight': 1,
    'eta': 0.01,
    'colsample_bytree': 0.5,
    'max_depth': 12,
    'subsample': 0.1,
    'reg_alpha': 1,
    'gamma': 0.05,
    'silent':True,
    "verbose_eval":10,
     "eval_metric":"mae",
    'seed': RANDOM_STATE
}

# xgtrain = xgb.DMatrix(X, label=loss_boxcox[0])
xgtrain = xgb.DMatrix(X_train, label=y_train_log)
xgval = xgb.DMatrix(X_test, label=y_test_log)
xgtest = xgb.DMatrix(X_test)
```

```python
gb_model = xgb.train(params, 
                     dtrain=xgtrain, 
                     evals=[(xgval,"validation")], 
                     early_stopping_rounds = 30,
                     num_boost_round = 2000)
```

# Prediction and evaluation on test

```python
gb_pred = gb_model.predict(xgb.DMatrix(X_test))

eva(gb_pred, y_test_log)
```

model

```python
def NN():
    model = Sequential()
    model.add(Dropout(0.1,input_shape = [X_train.shape[1],]))
    model.add(Dense(400,init = 'he_normal'))
    model.add(BatchNormalization())
    model.add(PReLU())
    model.add(Dropout(0.4))
    
    model.add(Dense(200, init = 'he_normal')) 
    model.add(BatchNormalization())  
    model.add(PReLU()) 
    model.add(Dropout(0.2))
    
    model.add(Dense(200, init = 'he_normal')) 
    model.add(BatchNormalization())  
    model.add(PReLU()) 
    model.add(Dropout(0.2))

    model.add(Dense(50, init = 'he_normal')) 
    model.add(BatchNormalization())
    model.add(PReLU())  
    
    model.add(Dense(1, init='he_normal', activation='sigmoid'))
    
    model.compile(loss='binary_crossentropy',metrics=['accuracy'], optimizer = "adadelta")
    return(model)
```
